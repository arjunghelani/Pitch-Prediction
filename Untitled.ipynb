```python
# Merge batter names with batter id's  

id_map = id_map.rename(columns={'MLBID':'batter'})
df['batter'] = df['batter'].astype('float')
df = df.merge(id_map[['batter', 'FANGRAPHSNAME']], on = 'batter').rename(columns={'FANGRAPHSNAME':'batter_name'})
```

# Feature Engineering 

```python
# Create a binary column if the batter swung on the pitch
swing_des = ['foul_tip', 'foul', 'swinging_strike', 'hit_into_play', 'swinging_strike_blocked', 'foul_bunt', \
             'bunt_foul_tip', 'missed_bunt']

df['swung'] = np.where(df['description'].isin(swing_des), 1, 0)
```

```python
# Group pitch types together

fastball = ['4-Seam Fastball', 'Sinker', 'Cutter']

curves = ['Curveball', 'Knuckle-Curve', 'Slow Curve', 'Slurve', 'Knuckleball']
sliders = ['Slider', 'Sweeper']
changeups = ['Changeup', 'Split-Finger', 'Euphus']

conditions = (
    (df['pitch_name'].isin(fastball)),
    (df['pitch_name'].isin(curves)),
    (df['pitch_name'].isin(sliders)),
    (df['pitch_name'].isin(changeups))
    
)

values = ['fastball', 'curveball', 'slider', 'off_speed']

df['gen_pitch_type'] = np.select(conditions, values)
```

```python
# Create a boolean column whether a pitch was a true ball/strike 

df = df[(df.plate_z.notna()) & (df.plate_x.notna())]

# '17/24' was calculated by taking the dimensions of home plate 

strike_conditions = (
    (df['plate_z'] <= df['sz_top']) & 
    (df['plate_z'] >= df['sz_bot']) & 
    (df['plate_x'] >= -17/24) & 
    (df['plate_x'] <= 17/14)
)

df['true_strike'] = np.where(strike_conditions, 1, 0)
```

```python
# Create a column for who the count favors

def count_conditions(s):
    if s['balls'] > s['strikes']:
        return('batter')
    elif s['balls'] < s['strikes']:
        return('pitcher')
    else:
        return('neutral')
    
df['count_favors'] = df.apply(count_conditions, axis=1)
```

```python
df['run_diff'] = df['fld_score'] - df['bat_score']
```

## Batter Feature Engineering

```python
# Filter out rows where general pitch type unknown
df = df[df.gen_pitch_type != '0']

# Group by batter and get how many pitches they faced of each type
# Pivot table to turn duplicate columns into rows
batter_pc = df.groupby(['batter', 'batter_name'], as_index=False).value_counts(['gen_pitch_type'])
batter_pc = batter_pc.pivot_table(index=['batter','batter_name'], columns='gen_pitch_type', values='count').fillna(0)

# Merge dataframe with sum of each row (total pitches)
batter_pc = batter_pc.sum(axis=1).to_frame().rename(columns={0:'total'}).\
            merge(batter_pc, left_index=True, right_index=True).reset_index()

# Filter out any batters with under 150 PA
pc_counts = batter_pc[['batter_name', 'total']]
pc_counts = pc_counts.loc[pc_counts.total > 150]
batter_pc = batter_pc.loc[batter_pc.batter_name.isin(pc_counts.batter_name)]
```

```python

# Filter out any pitches where the batter doesn't have 150 pitches faced 
df = df.loc[df.batter_name.isin(pc_counts.batter_name)]
```

```python
# Create df of percent of pitch type thrown for each batter
batter_pc_pct = batter_pc.iloc[:,3:].div(batter_pc.total, axis=0)
batter_pc_pct = pd.concat([batter_pc[['batter', 'batter_name', 'total']], batter_pc_pct], axis=1)

# Create merging dataframe with suffixes for later
batter_pc_pct_merger = pd.concat([
    batter_pc_pct[['batter', 'batter_name', 'total']], 
    batter_pc_pct.drop(columns=['batter', 'batter_name', 'total']).add_suffix('_pct_thrown')], axis=1)
```

```python
# Create boolean column if the batter chased
df['chase'] = np.where((df['swung'] == 1) & (df['true_strike'] == 0), True, False)

# Summarize and get number of chases by pitch for each batter
batter_chase_counts = df.groupby(['batter', 'batter_name','gen_pitch_type'], as_index=False).value_counts(['chase'])
batter_chase_counts = batter_chase_counts.pivot_table(index=['batter','batter_name', 'gen_pitch_type'], \
                                                      columns=['chase'], values='count').reset_index().\
                                                    rename(columns={False:'good', True:'bad'}).fillna(0)

batter_chase_counts['chase_pct'] = batter_chase_counts['bad'] / (batter_chase_counts['good'] + \
                                                                 batter_chase_counts['bad'])

# Percentage of pitches chased by pitch type for each batter
batter_chase_counts_pct = batter_chase_counts.pivot_table(index=['batter', 'batter_name'], columns='gen_pitch_type', values='chase_pct') \
                                                                                    .reset_index().fillna(0)

# Create merging df with suffixes to merge with other merging df
batter_chase_counts_pct_merger = pd.concat([
    batter_chase_counts_pct[['batter', 'batter_name']], 
    batter_chase_counts_pct.drop(columns=['batter', 'batter_name']).add_suffix('_pct_chased')], axis=1)

advanced_stats_batters = batter_pc_pct_merger.merge(batter_chase_counts_pct_merger, on=['batter', 'batter_name'])\
                                                                            .rename(columns={'total':'total_pitches'})
```

```python
# Create boolean column if the batter took a strike
df['take'] = np.where((df['swung'] == 0) & (df['true_strike'] == 1), True, False)

# Summarize and get number of takes by pitch for each batter
batter_take_counts = df.groupby(['batter', 'batter_name','gen_pitch_type'], as_index=False).value_counts(['take'])
batter_take_counts = batter_take_counts.pivot_table(index=['batter','batter_name', 'gen_pitch_type'], \
                                                      columns=['take'], values='count').reset_index().\
                                                    rename(columns={False:'good', True:'bad'}).fillna(0)

batter_take_counts['take_pct'] = batter_take_counts['good'] / (batter_take_counts['good'] + \
                                                                 batter_take_counts['bad'])

# Percentage of pitches taked by pitch type for each batter
batter_take_counts_pct = batter_take_counts.pivot_table(index=['batter', 'batter_name'], columns='gen_pitch_type', values='take_pct') \
                                                                                    .reset_index().fillna(0)

batter_take_counts_pct_merger = pd.concat([
    batter_take_counts_pct[['batter', 'batter_name']], 
    batter_take_counts_pct.drop(columns=['batter', 'batter_name']).add_suffix('_pct_taken')], axis=1)

advanced_stats_batters = advanced_stats_batters.merge(batter_take_counts_pct_merger, on=['batter', 'batter_name'])
```

```python
# Turn the batting ID of baserunners into binary 

df['on_1b'] = df['on_1b'].replace(np.nan, 0)
df['on_2b'] = df['on_2b'].replace(np.nan, 0)
df['on_3b'] = df['on_3b'].replace(np.nan, 0)

def baserunner_to_binary(idx):
    if idx != 0:
        return(1)
    else:
        return(0)
    
df['on_1b'] = df['on_1b'].apply(lambda x: baserunner_to_binary(x))
df['on_2b'] = df['on_2b'].apply(lambda x: baserunner_to_binary(x))
df['on_3b'] = df['on_3b'].apply(lambda x: baserunner_to_binary(x))

```

```python
# Estimated wOBA of each batter for each pitch type

est_woba = df.groupby(['batter', 'batter_name', 'gen_pitch_type'], as_index=False).agg({'estimated_woba_using_speedangle':'mean'}).\
sort_values('estimated_woba_using_speedangle', ascending=False)

est_woba = est_woba.pivot_table(index=['batter', 'batter_name'], columns='gen_pitch_type', \
                                values='estimated_woba_using_speedangle').fillna(0).reset_index()

est_woba_merger = pd.concat([
    est_woba[['batter', 'batter_name']], 
    est_woba.drop(columns=['batter', 'batter_name']).add_suffix('_est_woba')], axis=1)

advanced_stats_batters = advanced_stats_batters.merge(est_woba_merger, on=['batter', 'batter_name'])
```

```python
# Estimated batting average of each batter for each pitch type

est_ba = df.groupby(['batter', 'batter_name', 'gen_pitch_type'], as_index=False).agg({'estimated_ba_using_speedangle':'mean'}).\
sort_values('estimated_ba_using_speedangle', ascending=False)

est_ba = est_ba.pivot_table(index=['batter', 'batter_name'], columns='gen_pitch_type', \
                                values='estimated_ba_using_speedangle').fillna(0).reset_index()

est_ba_merger = pd.concat([
    est_ba[['batter', 'batter_name']], 
    est_ba.drop(columns=['batter', 'batter_name']).add_suffix('_est_ba')], axis=1)

advanced_stats_batters = advanced_stats_batters.merge(est_ba_merger, on=['batter', 'batter_name'])
```

```python
iso_value = df.groupby(['batter', 'batter_name', 'gen_pitch_type'], as_index=False).agg({'iso_value':'mean'}).\
                                                                                                        fillna(0)
iso_value = iso_value.pivot_table(index=['batter', 'batter_name'], columns='gen_pitch_type', values='iso_value').\
                                                                                                    reset_index ()

iso_value_merger = pd.concat([
    iso_value[['batter', 'batter_name']], 
    iso_value.drop(columns=['batter', 'batter_name']).add_suffix('_iso')], axis=1)

advanced_stats_batters = advanced_stats_batters.merge(iso_value_merger, on=['batter', 'batter_name'])
```

```python
hip_out = {'field_out', 'grounded_into_double_play', 'fielders_choice', 'force_out', 
           'sac_fly_double_play', 'fielders_choice_out', 'sac_fly', 'other_out', 'triple_play', 'field_error'}
hip_not_out = {'home_run', 'single', 'double', 'triple'}

strikeout = {'strikeout', 'strikeout_double_play'}

walk = {'walk'}

def Union(a,b,c,d):
    s1 = a.union(b)
    s2 = s1.union(c)
    s3 = s2.union(d)
    return s3

other = set(df.events.unique()) - Union(hip_out, hip_not_out, strikeout, walk)

events_conditions = (
    (df.events.isin(hip_out)),
    (df.events.isin(hip_not_out)),
    (df.events.isin(strikeout)),
    (df.events.isin(walk)),
    (df.events.isin(other))
)

events_values = ['hit_in_play_out', 'hit_in_play_not_out', 'strikeout', 'walk', 'other']

df['events'] = np.select(events_conditions, events_values)
```

```python
# At bat counter for the whole season per batter
ab_eachgame = df.groupby(['game_date', 'game_pk', 'batter', 'batter_name', 'at_bat_number'], as_index=False).agg({'events':'last'})
player_ab_counter = (ab_eachgame.groupby(['game_pk', 'batter', 'batter_name']).cumcount() + 1)
ab_eachgame = pd.concat([ab_eachgame, player_ab_counter], axis=1).rename(columns={0:'ab_counter'})

ab_eachgame['last_ab_outcome'] = ab_eachgame.groupby(['game_pk', 'batter', 'batter_name'])['events'].shift(1, axis=0)

df = df.merge(ab_eachgame.drop(columns='events'), on=['game_date', 'game_pk', 'batter', 'batter_name', 'at_bat_number'])
```

## Pitching Feature Engineering

```python
# Pitcher pitch mix breakdown 

pitcher_pc = df.groupby(['pitcher', 'player_name'], as_index=False).value_counts(['gen_pitch_type'])
pitcher_pc = pitcher_pc.pivot_table(index=['pitcher','player_name'], columns='gen_pitch_type', values='count').fillna(0)

pitcher_pc = pitcher_pc.sum(axis=1).to_frame().rename(columns={0:'total'}).\
            merge(pitcher_pc, left_index=True, right_index=True).reset_index()

pitcher_pc_counts = pitcher_pc[['player_name', 'total']]
pitcher_pc_counts = pitcher_pc_counts.loc[pitcher_pc_counts.total > 150]

pitcher_pc = pitcher_pc.loc[pitcher_pc.player_name.isin(pitcher_pc_counts.player_name)]

pitcher_pc_pct = pitcher_pc.iloc[:,3:].div(pitcher_pc.total, axis=0)
pitcher_pc_pct = pd.concat([pitcher_pc[['pitcher', 'player_name', 'total']], pitcher_pc_pct], axis=1)

pitcher_pc_pct_merger = pd.concat([
    pitcher_pc_pct[['pitcher', 'player_name', 'total']], 
    pitcher_pc_pct.drop(columns=['pitcher', 'player_name', 'total']).add_suffix('_pct_thrown')], axis=1)
```

```python
pitcher_chase_counts = df.groupby(['pitcher', 'player_name', 'gen_pitch_type'], as_index=False).value_counts(['chase'])
pitcher_chase_counts = pitcher_chase_counts.pivot_table(index=['pitcher', 'player_name', 'gen_pitch_type'], \
                                    columns='chase', values='count').reset_index().\
                                    rename_axis(None, axis=1).fillna(0).rename(columns={False:'bad', True:'good'})

pitcher_chase_counts['chase_pct'] = pitcher_chase_counts['bad'] / (pitcher_chase_counts['good'] + \
                                                                 pitcher_chase_counts['bad'])


pitcher_chase_counts_pct = pitcher_chase_counts.pivot_table(index=['pitcher', 'player_name'], columns='gen_pitch_type', values='chase_pct') \
                                                                                    .reset_index().fillna(0)

pitcher_chase_counts_pct_merger = pd.concat([
    pitcher_chase_counts_pct[['pitcher', 'player_name']], 
    pitcher_chase_counts_pct.drop(columns=['pitcher', 'player_name']).add_suffix('_pct_chased')], axis=1)

advanced_stats_pitchers = pitcher_pc_pct_merger.merge(pitcher_chase_counts_pct_merger, on=['pitcher', 'player_name'])\
                                                                            .rename(columns={'total':'total_pitches'})
```

```python
# Estimated wOBA against each pitcher for each pitch type

pitcher_est_woba = df.groupby(['pitcher', 'player_name', 'gen_pitch_type'], as_index=False).agg({'estimated_woba_using_speedangle':'mean'}).\
sort_values('estimated_woba_using_speedangle', ascending=False)

pitcher_est_woba = pitcher_est_woba.pivot_table(index=['pitcher', 'player_name'], columns='gen_pitch_type', \
                                values='estimated_woba_using_speedangle').fillna(0).reset_index()

pitcher_est_woba_merger = pd.concat([
    pitcher_est_woba[['pitcher', 'player_name']], 
    pitcher_est_woba.drop(columns=['pitcher', 'player_name']).add_suffix('_est_woba')], axis=1)

advanced_stats_pitchers = advanced_stats_pitchers.merge(pitcher_est_woba_merger, on=['pitcher', 'player_name'])
```

```python
# Estimated batting average of each pitcher for each pitch type

pitcher_est_ba = df.groupby(['pitcher', 'player_name', 'gen_pitch_type'], as_index=False).agg({'estimated_ba_using_speedangle':'mean'}).\
sort_values('estimated_ba_using_speedangle', ascending=False)

pitcher_est_ba = pitcher_est_ba.pivot_table(index=['pitcher', 'player_name'], columns='gen_pitch_type', \
                                values='estimated_ba_using_speedangle').fillna(0).reset_index()

pitcher_est_ba_merger = pd.concat([
    pitcher_est_ba[['pitcher', 'player_name']], 
    pitcher_est_ba.drop(columns=['pitcher', 'player_name']).add_suffix('_est_ba')], axis=1)

advanced_stats_pitchers = advanced_stats_pitchers.merge(pitcher_est_ba_merger, on=['pitcher', 'player_name'])
```

```python
pitcher_iso_value = df.groupby(['pitcher', 'player_name', 'gen_pitch_type'], as_index=False).agg({'iso_value':'mean'}).\
                                                                                                        fillna(0)
pitcher_iso_value = pitcher_iso_value.pivot_table(index=['pitcher', 'player_name'], columns='gen_pitch_type', values='iso_value').\
                                                                                                    reset_index().fillna(0)

pitcher_iso_value_merger = pd.concat([
    pitcher_iso_value[['pitcher', 'player_name']], 
    pitcher_iso_value.drop(columns=['pitcher', 'player_name']).add_suffix('_iso')], axis=1)

advanced_stats_pitchers = advanced_stats_pitchers.merge(pitcher_iso_value_merger, on=['pitcher', 'player_name'])
```

```python
# Create a last pitch column

df['last_pitch'] = df.gen_pitch_type.shift(1, axis=0)

lp_conditions = ((df['balls'] == 0) & (df['strikes'] == 0))

df['last_pitch'] = np.where(lp_conditions, None, df['last_pitch'])

# Create last pitch velocity column

df['last_pitch_velo'] = df.effective_speed.shift(1, axis=0)

lp_conditions = ((df['balls'] == 0) & (df['strikes'] == 0))

df['last_pitch_velo'] = np.where(lp_conditions, None, df['last_pitch_velo'])
```

```python
model_df = df[['gen_pitch_type', 'batter', 'batter_name', 'pitcher', 'player_name', 'stand', 
               'p_throws', 'balls', 'strikes', 'on_3b', 'on_2b', 'on_1b', 'at_bat_number', 'pitch_number',
               'count_favors', 'ab_counter', 'last_ab_outcome', 'last_pitch_velo', 'run_diff']]
```

## Model Preparation

```python
advanced_stats_batters_merger = advanced_stats_batters.iloc[:, :2].join(advanced_stats_batters.iloc[:, 3:].\
                                                                 add_suffix('_batter'))

advanced_stats_pitchers_merger = advanced_stats_pitchers.iloc[:, :2].join(advanced_stats_pitchers.iloc[:, 3:].\
                                                                 add_suffix('_pitcher'))

model_df = model_df.merge(advanced_stats_batters_merger, on=['batter', 'batter_name'])
model_df = model_df.merge(advanced_stats_pitchers_merger, on=['pitcher', 'player_name'])
```

```python
model_df['stand'] = model_df['stand'].astype('category')
model_df['p_throws'] = model_df['p_throws'].astype('category')
model_df['count_favors'] = model_df['count_favors'].astype('category')
model_df['last_ab_outcome'] = model_df['last_ab_outcome'].astype('category')

model = pd.get_dummies(data=model_df, columns=['stand', 'p_throws', 'count_favors', 'last_ab_outcome'], \
                        dummy_na=True)

model['last_pitch_velo'] = model['last_pitch_velo'].fillna(0)
```

## Random Forest

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
target = model['gen_pitch_type']
features = model.iloc[:, 5:]

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2)

rf_clf = RandomForestClassifier()
```

```python
%%time

rf_clf.fit(X_train, y_train)
```

```python
rf_clf.score(X_test, y_test)
```
